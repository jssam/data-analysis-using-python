Load the train and test dataset using pandas Library  

```python
train_data=pd.read_csv('train.csv')
test_data=pd.read_csv('test (1).csv')
```

Let us perform a Sanity check to make sure that Dataset is loaded

### Analyzing the age column in the given dataset
- It is observed that the age columns has many missing values
- To solve this we find mean of all the ages and replace missing value with the mean

```python
mean_age=train_data['Age'].mean()
train_data['Age'].replace(np.nan,mean_age)
```

```python
train_data['Age']=train_data['Age'].replace(np.nan,mean_age)
train_data.head(6)
```

The updated values of Age are in Decimeal form so we round off this value using the .round() function of pandas

```python
#rounding off ages
train_data['Age']=train_data['Age'].round()
train_data.head(8)
```

Let us rename the column Pclass to Ticketclass

```python
#renaming the Pclass columns of dataset
train_data=train_data.rename(columns={'Pclass':'Ticketclass'})
test_data=test_data.rename(columns={'Pclass':'Ticketclass'})
```

```python
print(train_data)
```

## One Hot Encoding
- This is a standard procedure which is commonly used in all types of Datasets
- This is done in order to change categorical variables to numerical variables,since computers cannot understand categorical variable well,for this specific case
- Pandas already has a built in functionality .get_dummies() to make our task of encoding easier

```python
#one hot encoding
train_data=pd.get_dummies(train_data,columns=['Sex','Embarked'])
test_data=pd.get_dummies(test_data,columns=['Sex','Embarked'])
```

```python
test_data.head()
```

```python
train_data.head()
```

# Dropping Unnecessary Data
While doing analysis of data not every column is meaningful for our work, adding unwanted data for analysis can skew our 
results and we mightend up spending majority of our time in finetuning our model since it did not perform well due to unwanted features

```python
train_data = train_data.drop(['Name','Ticket','Fare','Cabin'],axis =1)
train_data.head()
```

```python
train_data.head()
```

```python
test_data= test_data.drop(['Name','Ticket','Fare','Cabin',],axis =1)

test_data.head()
```

# Rearranging the Data 
- This is the final step of Data-preprocessing
- This is not necessary, but this step makes our work more simpler and easier

```python
data_train = train_data[['PassengerId','Sex_female','Sex_male','SibSp','Embarked_C','Embarked_S','Embarked_Q','Parch','Ticketclass','Survived']]
data_test = test_data[['PassengerId','Sex_female','Sex_male','Embarked_C','Embarked_S','Embarked_Q','SibSp','Parch','Ticketclass']]
data_train.head()
```

# Feeding and Normalizing Data
- We need to two Variables X and Y for this purpose
- Since Our Pandas Dataframe has total 9 features, we put these 9 features into X variable 
- The last feature is the Ground Truth i.e 'Survived' we take this in a seperate variable Y, Since this is what we are trying to predict
- In order to Normalize the Data we use the Standard Scalar function of sklearn library, which adds bias to the data

```python
X_train=data_train.iloc[:,0:9]
y_train=data_train.iloc[:,9]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(data_test)
```

# The Deep Learning Part
- Import the neccary packages of keras library
- we have chosen 9 features at the start of X_train so our Neural Network should have 9 input neurons
- The output is the Survival Rate which has only two possibilties either "true or false" i.e "0 or 1" we have 2 output neurons at the last layer with Sigmoid Activation
- Let us use a relatively shallow neural network with relu activation function in the starting layers, Since our Dataset is small and Relu acts as a Non-Linear activation function

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.layers import LeakyReLU
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(output_dim = 7, init = 'uniform', activation = 'relu', input_dim = 9))


#Hidden layer
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu'))
classifier.add(Dense(output_dim = 4, init = 'uniform', activation = 'relu'))

classifier.add(Dense(output_dim = 2, init = 'uniform', activation = 'relu'))

#output layer with 1 output neuron which will predict 1 or 0
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))
```

```python
classifier.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size = 10,validation_split=0.3, nb_epoch = 50)
```

```python
#getting predictions of test data
data_check=pd.read_csv('gender_submission.csv')
prediction = classifier.predict(X_test).tolist()
# list to series
se = pd.Series(prediction)
# creating new column of predictions in data_check dataframe
data_check['check'] = se
data_check['check'] = data_check['check'].str.get(0)
```

```python
data_check['check']=data_check['check'].round()
data_check.head()
```
